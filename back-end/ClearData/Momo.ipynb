# Merchants
Một số mẫu thông tin của thương nhân MoMo.
* SERVICE_ID: Mã định danh duy nhất cho dịch vụ. Mỗi thương gia có thể có nhiều ID dịch vụ cho các công ty con và cửa hàng của mình.
* STORE_ID: Một định danh duy nhất cho một cửa hàng. Một thương gia có thể có nhiều cửa hàng tại các địa điểm khác nhau.
* MERCHANT_NAME: Tên của thương gia. Lưu ý rằng một thương gia có thể có nhiều tên nhưng tất cả chúng đều được đăng ký theo cùng một DỊCH VỤ_ID
* MERCHANT_ADDRESS: Địa chỉ của thương gia, thường là trụ sở chính.
* STORE_NAME: Tên của cửa hàng.
* STORE_ADDRESS: Địa chỉ của cửa hàng.

# Transactions
Một số mẫu giao dịch MoMo.
* TRANSACTION_ID: ID duy nhất cho giao dịch.
* USER_ID: ID duy nhất cho người dùng.
* SERVICE_ID: Dịch vụ mà người dùng đã trả cho giao dịch.
* STORE_ID: ID của cửa hàng giao dịch.
* TRANSACTION_AMOUNT: Số tiền của giao dịch.
* USER_BANK: ID ngân hàng của người dùng tại thời điểm người dùng thực hiện giao dịch.

```python
# Tìm ra các id transaction bị trùng lặp
duplicate_tran_id = tra["TRANSACTION_ID"].value_counts().reset_index(name="Duplicate_Count")
duplicate_tran_id = duplicate_tran_id[duplicate_tran_id["Duplicate_Count"] > 1]
duplicate_tran_id = [i for i in duplicate_tran_id['index']]
duplicate_tran_id
```

```python
# Danh sách các cửa hàng bị trùng lặp
tra[tra["TRANSACTION_ID"].apply(lambda id: True if id in duplicate_tran_id else False)]
```

```python
# Lấy ra những phần tử bị trùng lặp ra một phần tử
def Remove_Dup(id):
    if id in duplicate_tran_id:
        duplicate_tran_id.remove(id)
        return True
    else:
        return False
tran_dup = tra[tra["TRANSACTION_ID"].apply(Remove_Dup)]
tran_dup
```

```python
# Xóa bớt các transaction bị trùng lặp
tra = tra.drop(tran_dup.index)
tra
```
