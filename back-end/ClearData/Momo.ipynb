# Merchants
Một số mẫu thông tin của thương nhân MoMo.
* SERVICE_ID: Mã định danh duy nhất cho dịch vụ. Mỗi thương gia có thể có nhiều ID dịch vụ cho các công ty con và cửa hàng của mình.
* STORE_ID: Một định danh duy nhất cho một cửa hàng. Một thương gia có thể có nhiều cửa hàng tại các địa điểm khác nhau.
* CATEGORY: loại cửa hàng của store
* MERCHANT_NAME: Tên của thương gia. Lưu ý rằng một thương gia có thể có nhiều tên nhưng tất cả chúng đều được đăng ký theo cùng một DỊCH VỤ_ID
* MERCHANT_ADDRESS: Địa chỉ của thương gia, thường là trụ sở chính.
* STORE_NAME: Tên của cửa hàng.
* STORE_ADDRESS: Địa chỉ của cửa hàng.

```python
mer["category"] = mer["category"].apply(lambda cat: cat.lower())
print("Có",mer["category"].nunique(),"category khác nhau.")
mer["category"].unique()
```

```python
mer['key'] = mer[["service_id", "store_id"]].apply(lambda x: str(x['service_id']) + '-' + str(x['store_id']),axis=1)
mer.sample(10)
```

```python
duplicate_store = mer["key"].value_counts().reset_index(name="duplicate_count")
duplicate_store = duplicate_store[duplicate_store["duplicate_count"] > 1]
duplicate_store = [i for i in duplicate_store['index']]
print(mer[mer["key"].apply(lambda id: True if id in duplicate_store else False)].shape)
print(mer[~mer["key"].apply(lambda id: True if id in duplicate_store else False)].shape)
mer = mer[~mer["key"].apply(lambda id: True if id in duplicate_store else False)]
mer.sample(10)
```

```python
MOMO = mer["store_address"].str.contains(r'^MOMO', regex=True)
TCH = mer["store_address"].str.contains(r'^TCH', regex=True)
MINISTOPVN = mer["store_address"].str.contains(r'^MINISTOPVN', regex=True)
HOANGYENBUFFET = mer["store_address"].str.contains(r'^HOANGYENBUFFET', regex=True)
dulieuhople = mer[~MOMO & ~TCH & ~MINISTOPVN & ~HOANGYENBUFFET]
dulieuhople["category"].unique()
```
